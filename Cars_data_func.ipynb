```python
import pandas as pd
import numpy as np

df = pd.read_csv('Input.csv',skiprows = 2)
df.head()
```

```python
def drop_duplicates(data,col1):
    df.drop_duplicates(subset = col1,keep = False, inplace = True)
    return df

drop_duplicates(df,'Model')
```

```python
def data_replace(data,col1):
    data[col1] = data[col1].replace('[\$\,]',"",regex=True)
    return data

data_replace(df,'Target')
```

```python
def data_convert(data,col1,dtype):
    data[col1] = data[col1].astype(dtype)
    return data

data_convert(df,'Target','float')
```

```python

def data_calc(data,col1,val1):
    data[col1]=data[col1].div(val1)
    return data

data_calc(df,'Sales',10)
```

```python
def data_sort(data,col1):
    data.sort_values(by=[col1],ascending=True)
    return data

data_sort(df,'Model')

```

```python
def data_status_calc(data,col1,col2,val1,val2):
    data[col2] = data[col2].astype(float)
    data[col1] = [val1 if a < 1000 else val2 for a in data[col2]]
    return data

data_status_calc(df,'Status','Target','Alert','Ideal')
```
